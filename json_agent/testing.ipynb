```python
theorem = """ ### The Banach-Stone Theorem

**Theorem:** Let $X$ and $Y$ be compact Hausdorff spaces. Let $C(X)$ and $C(Y)$ be the Banach spaces of continuous, real-valued functions on $X$ and $Y$, equipped with the supremum norm. If there exists a surjective linear isometry $T: C(X) \to C(Y)$, then $X$ and $Y$ are homeomorphic.

---

### Proof

**Step 1: The Adjoint Operator**
Because $T: C(X) \to C(Y)$ is a surjective linear isometry, its adjoint operator $T^*: C(Y)^* \to C(X)^*$ is also a surjective linear isometry between the continuous dual spaces. Furthermore, $T^*$ is continuous with respect to the weak-* topologies on both dual spaces. 

**Step 2: Preservation of Extreme Points**
Let $B_X$ and $B_Y$ denote the closed unit balls of $C(X)^*$ and $C(Y)^*$. Because $T^*$ is a linear isometry, it maps $B_Y$ bijectively onto $B_X$. Since $T^*$ is affine and bijective, it must map the set of extreme points of $B_Y$ bijectively onto the extreme points of $B_X$:
$$T^*(\text{Ext}(B_Y)) = \text{Ext}(B_X)$$

**Step 3: Characterizing the Extreme Points**
By the Riesz-Markov-Kakutani representation theorem, the dual space $C(X)^*$ is isometrically isomorphic to the space of regular signed Borel measures on $X$ with the total variation norm. The extreme points of the unit ball of this measure space are precisely the scaled Dirac evaluation functionals. Thus, for any compact Hausdorff space $K$:
$$\text{Ext}(B_K) = \{\pm e_k \mid k \in K\}$$
where $e_k(f) = f(k)$. 

**Step 4: Inducing a Bijection**
Since $T^*$ maps extreme points to extreme points, for every $y \in Y$, there exists a unique $x \in X$ and a sign $s(y) \in \{-1, 1\}$ such that:
$$T^*(e_y) = s(y) e_x$$
We define a map $\varphi: Y \to X$ by setting $\varphi(y) = x$. Because $T^*$ is a bijection on the extreme points, it is straightforward to verify that $\varphi$ is a bijection.

**Step 5: Extracting the Homeomorphism**
By the definition of the adjoint and the evaluation functional, for any $f \in C(X)$:
$$T^*(e_y)(f) = e_y(Tf) = (Tf)(y)$$
Equating this with our extreme point mapping yields:
$$(Tf)(y) = s(y) f(\varphi(y))$$
If we evaluate this using the constant function $\mathbf{1} \in C(X)$, we get:
$$(T\mathbf{1})(y) = s(y) \mathbf{1}(\varphi(y)) = s(y)$$
Since $T\mathbf{1} \in C(Y)$, the sign function $s(y)$ is a continuous function from $Y$ to $\{-1, 1\}$. 

**Step 6: Continuity of $\varphi$**
We can rewrite our equation as $f(\varphi(y)) = s(y) (Tf)(y)$. Since $s(y)$ and $(Tf)(y)$ are both continuous functions on $Y$, their product is continuous. Therefore, the composition $f \circ \varphi$ is continuous for every $f \in C(X)$. Because $X$ is a compact Hausdorff space, it is completely regular—meaning its topology is exactly the weak topology generated by $C(X)$. Since $f \circ \varphi$ is continuous for all $f$, the map $\varphi: Y \to X$ itself must be continuous. 

A continuous bijection between compact Hausdorff spaces is a homeomorphism, completing the proof.
"""
```
